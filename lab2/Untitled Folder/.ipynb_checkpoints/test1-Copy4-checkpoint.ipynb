```python
# Joo Kai Tay, 22489437, lab02
# Importing libraries 
import numpy as np
import math
import matplotlib.pyplot as plt
from skimage import io, color
from scipy import signal 
import cv2
```

## Load the original images

```python
# Read Image 1
file_name_1 = 'cat.jpg'
im1 = cv2.imread(file_name_1)

# Convert Original Image 1 to grayscale 
gray1 = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)

# Read Image 2
file_name_2 = 'mouse.jpg'
im2 = cv2.imread(file_name_2)

# Convert Original Image 1 to grayscale 
gray2 = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)

# Display Images
plt.subplot(1,2,1), plt.imshow(cv2.cvtColor(gray1, cv2.COLOR_BGR2RGB)), plt.title("Original Image of: " + file_name_1)
plt.subplot(1,2,2), plt.imshow(cv2.cvtColor(gray2, cv2.COLOR_BGR2RGB)), plt.title("Original Image of: " + file_name_2)
```

## Convert images into the frequency domain and apply low and high pass filters

```python
# Apply fourier transform to change images into frequency domain
freq_img1 = np.fft.fft2(gray1)
freq_img2 = np.fft.fft2(gray2)

ftimage1 = np.fft.fftshift(freq_img1)
ftimage2 = np.fft.fftshift(freq_img2)

plt.subplot(1,2,1), plt.imshow(np.log(np.abs(ftimage1)), cmap='gray'), plt.title("Fourier Transform of: " + file_name_1)
plt.subplot(1,2,2), plt.imshow(np.log(np.abs(ftimage2)), cmap='gray'), plt.title("Fourier Transform of: " + file_name_2)
```
