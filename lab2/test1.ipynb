## Load the original images

```python
# Read Image 1
file_name_1 = 'cat.jpg'
im1 = cv2.imread(file_name_1)

# Convert Original Image 1 to grayscale 
gray1 = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)

# Read Image 2
file_name_2 = 'mouse.jpg'
im2 = cv2.imread(file_name_2)

# Convert Original Image 1 to grayscale 
gray2 = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
```

## Convert images into the frequency domain and apply low and high pass filters

```python
# Define Butterworth low pass filter
def low_pass_filter(image, cutoff_frequency, sample_rate):
    M,N = image.shape
    H = np.zeros((M,N), dtype=np.float32)
    for u in range(M):
        for v in range(N):
            D = np.sqrt((u-M/2)**2 + (v-N/2)**2)
            H[u,v] = 1 / (1 + (D/cutoff_frequency)**sample_rate)
    return H
```

```python
# Define Butterworth high pass filter
def highpass(image, cutoff_freq, fs):
    M,N = image.shape
    HPF = np.zeros((M,N), dtype=np.float32)
    for u in range(M):
        for v in range(N):
            D = np.sqrt((u-M/2)**2 + (v-N/2)**2)
            if D != 0:
                HPF[u,v] = 1 / (1 + (cutoff_freq/D)**fs)
            else:
                HPF[u,v] = 0
    return HPF
```

```python
# Apply fourier transform to change images into frequency domain
freq_img1 = np.fft.fft2(gray1)
freq_img2 = np.fft.fft2(gray2)
ftimage1 = np.fft.fftshift(freq_img1)
ftimage2 = np.fft.fftshift(freq_img2)

# Generate filters
filter_1 = low_pass_filter(ftimage1, 10, 7)
filter_2 = highpass(ftimage2, 27, 4)

# Apply filters
filtered1 = ftimage1 * filter_1
filtered2 = ftimage2 * filter_2
```

```python
# Inverse the fourier transform to return to spatial domain
smooth = np.fft.ifftshift(filtered1)
smooth1 = np.abs(np.fft.ifft2(smooth))
smooth1 -= smooth1.min()
smooth1 = smooth1*255 / smooth1.max()
smooth1 = smooth1.astype(np.uint8)

sharp = np.fft.ifftshift(filtered2)
sharp2 = np.abs(np.fft.ifft2(sharp))
sharp2 -= sharp2.min()
sharp2 = sharp2*255 / sharp2.max()
sharp2 = sharp2.astype(np.uint8)
```

## Add the two images together

```python
# Add the frequencies together
combined_freq = filtered1 + filtered2

# Take it back to spatial domain
combined = np.fft.ifftshift(combined_freq)
combined1 = np.abs(np.fft.ifft2(combined))
combined1 -= combined1.min()
combined1 = combined1*255 / combined1.max()
combined1 = combined1.astype(np.uint8)
```

## Display Images 

```python
# Display all images as per specification
# Hybrid figure also displayed seperately below 
plt.figure(figsize=(10,10), constrained_layout=False)
plt.subplots_adjust(top=1.0)

plt.subplot(3,3,1), plt.imshow(cv2.cvtColor(gray1, cv2.COLOR_BGR2RGB)), plt.title("Original " + file_name_1)
plt.subplot(3,3,2), plt.imshow(np.log(np.abs(filtered1)), cmap='gray'), plt.title("Spectrum of " + file_name_1)
plt.subplot(3,3,3), plt.imshow(smooth1, cmap='gray'), plt.title("Smoothed " + file_name_1)


plt.subplot(3,3,4), plt.imshow(cv2.cvtColor(gray2, cv2.COLOR_BGR2RGB)), plt.title("Original " + file_name_2)
plt.subplot(3,3,5), plt.imshow(np.log(np.abs(filtered2) + 1e-3), cmap='gray'), plt.title("Spectrum of " + file_name_2)
plt.subplot(3,3,6), plt.imshow(sharp2, cmap='gray'), plt.title("Sharpened " + file_name_2)

plt.subplot(3,3,8), plt.imshow(np.log(np.abs(combined_freq)), cmap='gray'), plt.title("Spectrum of Hybrid Image")
plt.subplot(3,3,9), plt.imshow(combined1, cmap='gray'), plt.title("Hybrid Image")
```
